#1. Law of Large Numbers

#2. Central Limit Theorem

Consider a sequence of $n$ i.i.d. random variables $X_1, ..., X_n$, each with mean $\mu$ and variance $\sigma^2$. Define
$$ S_n = \frac{X_1 + X_2 + ... + X_n}{n}$$
then, the Central Limit Theorem states that, as $n\to\infty$,
$$Z_n \stackrel{\rm def}{=} \sqrt{n}\frac{S_n - \mu}{\sigma}\sim\mathcal{N}(0,1)$$

Given two arrays of coordinates $X$, $Y$, and a function $f(x,y, ...)$, the following function plots the points $(x,y)\in X\times Y$, coloring them according to the value of $f(x,y, ...)$.

```python
def scatter_grid(X, Y, f, *args, colorbar_label = None, **kwargs): #*args and **kwargs represent respectively additional positional and keyword arguments of the function f
  X_grid, Y_grid = np.meshgrid(X, Y)
  X_flat = X_grid.flatten()
  Y_flat = Y_grid.flatten()
  scatter = plt.scatter(X_flat, Y_flat, c = f(X_flat, Y_flat, *args, **kwargs), cmap = 'viridis')
  plt.colorbar(scatter, label=colorbar_label)
```

# 4. LLN and matrices


Consider a matrix $X\in\mathbb{R}^{n\times d}$ having i.i.d. standard Gaussian elements $X_{ij}$, and the correspondent *sample covariance matrix* $\Sigma = \frac{1}{n}X^TX$. The Law of Large Numbers, applied to each element of $\Sigma$, implies, as $n\to\infty$,
$$\Sigma_{ii} = \frac{1}{n}\sum_{k = 1}^n X_{ki}^2\to 1$$
and
$$ \Sigma_{ij} = \frac{1}{n}\sum_{k = 1}^n X_{ki}X_{kj}\to 0, \text{ for }i\neq j.$$

From this, one might conclude that $\Sigma \to\mathbb{I}_d$ as $n,d\to\infty$. However will show that the situation is more complex when we consider the matrix as a whole.
